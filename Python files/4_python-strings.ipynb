## split() method

# Regular expressions

Regular expressions provide a flexible way to search or match string patterns in text
<br>Python’s built-in re module is responsible for applying regular expressions to strings.
<br>The re module functions fall into three categories: pattern matching, substitution, and splitting

```python
import re
```

```python
text="Meier Maier Maayer MYER mYER"
```

Square brackets, "[" and "]", are used to include a character class.
<br>[xyz] means e.g. either an "x", an "y" or a "z".
<br>[a-e] a simplified writing for [abcde] or [0-5] denotes [012345].

```python
pattern=r"M[ae][iy]er"
```

```python
result=re.findall(pattern,text)
```

```python
result
```

# Other functions

Re.sub() finds all substrings where the RE matches, and replace them with a different string

```python
re.sub(r"a","*","ababababa")
```

Re.split() Split the string into a list, splitting it wherever the RE matches

```python
re.split(r"a","ababababa")
```

# regular expression rules

• \d matches any digit, and \D matches any non-digit

```python
result=re.findall(pattern="\d",string="AB.C 34 ABDC 45")
```

```python
result
```

```python
result=re.findall(pattern="\d+",string="AB.C 34 ABDC 45")
```

```python
result
```

\w matches any digit, w matches any letter or number, and \W matches anything else

```python
result=re.findall(pattern="\w",string="AB.C 34 ABDC 45")
```

```python
result
```

```python
result=re.findall(pattern="\w+",string="AB.C 34 ABDC 45")
```

```python
result
```

only find words having two characters

```python
result=re.findall(pattern="",string="AB.C 34 ABDC 45")
```

```python
result
```

• \s matches whitespace, and \S matches non-whitespace

```python
result=re.findall(pattern="\s",string="This is a test.\n it has whitespaces")
```

```python
result
```

```python
result=re.findall(pattern="\S",string="This is a test.\n it has whitespaces")
```

```python
result
```

```python
result=re.findall(pattern="\S+",string="This is a test.\n it has whitespaces")
```

```python
result
```
